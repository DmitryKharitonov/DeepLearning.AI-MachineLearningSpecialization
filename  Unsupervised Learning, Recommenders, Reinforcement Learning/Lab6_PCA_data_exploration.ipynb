# Solution

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
x = pd.read_csv("./Lab6_data.csv", delimiter=",")
x
```

```python
pca_2 = PCA(n_components=2)
x_2d = pca_2.fit_transform(x)
x_2d_reversed = pca_2.inverse_transform(x_2d)

fig, ax = plt.subplots(figsize=(3,3))
ax.scatter(x_2d[:,0], x_2d[:,1])
```

```python
pca_3 = PCA(n_components=3)
x_3d = pca_3.fit_transform(x)
x_3d_reversed = pca_3.inverse_transform(x_3d)

fig, ax = plt.subplots(figsize=(3,3))
ax.scatter(x_3d[:,0], x_3d[:,1], c=x_3d[:,2])
```
