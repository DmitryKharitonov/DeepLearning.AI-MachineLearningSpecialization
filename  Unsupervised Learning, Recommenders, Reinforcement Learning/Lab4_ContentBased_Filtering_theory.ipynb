To predict User's rating we should apply similar formula as in collaborative learning:

$x_i \cdot w_j + \cancel{b_j}$, where

* $x_i$ and $w_j$ are films attributes and user's preferences (compatible sizes)
* bias is removed

However, features sizes are not compatible, therefore we should convert them to required sized using NN:

* $X_m$ -> $V_m$
* $X_u$ -> $V_u$
* $V_m$ and $V_u$ are compatible sizes



For example:

$X_u$ -> $V_u$
* $X^0_u$ size is 128
* $X^1_u$ becomes 64
* $X^2_u$ becomes $V_u$ <b>32</b>

$X_m$ -> $V_m$
* $X^0_m$ size is 256
* $X^1_m$ becomes 128
* $X^2_m$ becomes $V_m$ with size <b>32</b>

Then, the prediction will work, conditional amount of rows is the same:

$$ Y_{pred} = V_m \cdot V_u$$

Cost function and everything else becomes same as in collaborative filtering
